##Data Cleaning
1. Handle missing or invalid values using appropriate methods (e.g., imputation,deletion).
2. Address inconsistencies in categorical columns (eg., typos, inconsistent capitalization).
3. Ensure numerical columns are in the correct format and free from outliers, where applicable.

```python
#Drop completely missing column
df_cleaned = df.drop(columns=['CAMPAIGN_NBR'], errors='ignore')
df_cleaned.info()
```

```python
# Missing values in categorical columns with mode
categorical_cols = df_cleaned.select_dtypes(include=['object']).columns
df_cleaned[categorical_cols] = df_cleaned[categorical_cols].fillna(df_cleaned[categorical_cols].mode().iloc[0])
df_cleaned[categorical_cols].isnull().sum()
```

```python
# Missing values in numerical columns with median
numerical_cols = df_cleaned.select_dtypes(include=['number']).columns
df_cleaned[numerical_cols] = df_cleaned[numerical_cols].fillna(df_cleaned[numerical_cols].median())
```

```python
# Date columns to proper datetime format
if 'REPAIR_DATE' in df_cleaned.columns:
    df_cleaned['REPAIR_DATE'] = pd.to_datetime(df_cleaned['REPAIR_DATE'], errors='coerce')
```

```python
# Converting cost-related columns to numeric
cost_columns = ['REPORTING_COST', 'TOTALCOST', 'LBRCOST']
for col in cost_columns:
    if col in df_cleaned.columns:
        df_cleaned[col] = pd.to_numeric(df_cleaned[col], errors='coerce')
```

```python
# Handle outliers in numerical columns
# df_cleaned[col].quantile(0.99) gets the value below which 99% of data points fall
for col in numerical_cols:
    upper_limit = df_cleaned[col].quantile(0.99)
    df_cleaned[col] = df_cleaned[col].clip(upper=upper_limit)
```

```python
text_data = df['GLOBAL_LABOR_CODE_DESCRIPTION'].dropna()  # Remove NaNs
```

```python
#Text Preprocessing
def preprocess_text(text):
    text = text.lower()  # Convert to lowercase
    text = re.sub(r'[^a-z\s]', '', text)  # Remove special characters
    words = word_tokenize(text)  # Tokenization
    words = [word for word in words if word not in stopwords.words('english')]  # Remove stopwords
    return words
```

```python
df['cleaned_text'] = text_data.apply(preprocess_text)
```

```python
all_words = [word for sublist in df['cleaned_text'] for word in sublist]
```

```python
common_tags = Counter(all_words).most_common(10)
print("10 Most Common Tags:", common_tags)
```

##Why Use TF-IDF Instead of Just Word Count?
- TF (Term Frequency) counts how often a word appears in a document.
IDF (Inverse Document Frequency) reduces the importance of common words that appear in many documents.
- TF-IDF balances both, giving more weight to words that are unique and informative.

```python
#TF-IDF Vectorization for Important Terms
vectorizer = TfidfVectorizer(max_features=10)

tfidf_matrix = vectorizer.fit_transform(text_data.dropna())
important_terms = vectorizer.get_feature_names_out()

print("Top TF-IDF Terms: ", important_terms)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np

df["ALL_TEXT"] = df["CORRECTION_VERBATIM"] + " " + df["CUSTOMER_VERBATIM"]

# TF-IDF to identify most common failure-related words
vectorizer = TfidfVectorizer(stop_words='english', max_features=20)
tfidf_matrix = vectorizer.fit_transform(df["ALL_TEXT"])


feature_names = np.array(vectorizer.get_feature_names_out())
tfidf_sums = np.array(tfidf_matrix.sum(axis=0)).flatten()
top_keywords = feature_names[np.argsort(-tfidf_sums)][:10]

top_keywords
```
